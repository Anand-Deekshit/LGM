```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
KOI = pd.read_csv('KOI.csv')
KOI.head()
```

```python
KOI = KOI[KOI['koi_disposition'] != 'FALSE POSITIVE']
KOI.head()
```

```python
KOI.columns
```

```python
relavent_cols = ['koi_period', 'koi_impact',
                 'koi_duration', 'koi_depth', 'koi_ror',
                 'koi_srho', 'koi_prad', 'koi_sma', 'koi_incl', 'koi_teq',
                 'koi_steff', 'koi_slogg', 'koi_smet', 'koi_srad',
                 'koi_smass']

KOI = KOI[relavent_cols]
KOI.head()
```

Fill NaNs with mean value of the columns

```python
for col in KOI.columns:
  mean_value = KOI[col].mean()
  KOI[col].fillna(value=mean_value, inplace=True)
KOI.head()
```

```python
X = KOI.drop('koi_prad', axis=1)
y = KOI['koi_prad']
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.model_selection import StratifiedKFold
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

gbm = xgb.XGBRegressor()
reg_cv = GridSearchCV(gbm, {'colsample_bytree': [1.0], 'min_child_weight':[1.0, 1.2],
                      'max_depth': [3, 4, 6], 'n_estimators':[500, 1000]}, verbose=1)

reg_cv.fit(X_train, y_train)

```

```python
reg_cv.best_params_
```

```python
gbm = xgb.XGBRegressor(**reg_cv.best_params_)
gbm.fit(X_train, y_train)
```

```python
predictions = gbm.predict(X_test)
predictions
```

```python
gbm.score(X_train, y_train)
```
